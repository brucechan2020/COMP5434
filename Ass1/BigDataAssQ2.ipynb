```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
```

```python
data = [[1, 2, 3], [2, 3, 4], [3, 4, 5], [4, 5, 6],[1, 3, 5],\
[2, 4, 6], [1, 3, 4], [2, 4, 5],[3, 5, 6], [1, 2, 4], [2, 3, 5], [3, 4, 6]]
te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = fpgrowth(df, min_support= 4/12, use_colnames=True)
frequent_itemsets
```

```python
def _hash(i, j):
    return (i * j) % 11
# Initialize the dictionary to store the counts of item pairs
item_pair_counts = {}

b = 11
hash_table = {i: 0 for i in range(b)}

# Iterate through each transaction and count the frequency of each item pair
for transaction in data:
    for i in range(len(transaction)):
        for j in range(i+1, len(transaction)):
            item_pair = frozenset([transaction[i], transaction[j]])
            bucket = _hash(transaction[i], transaction[j])
            hash_table[bucket] += 1
            if item_pair in item_pair_counts:
                item_pair_counts[item_pair] += 1
            else:
                item_pair_counts[item_pair] = 1

# Print the counts of all item pairs
for item_pair, count in item_pair_counts.items():
    print(f"Item pair {tuple(item_pair)}: {count} ; basker no: {_hash(tuple(item_pair)[0],tuple(item_pair)[1])}")

print("----------------------------------")
# Print the counts in each bucket
for bucket, count in hash_table.items():
    print(f"Bucket {bucket}: {count}")

print("----------------------------------")
print("keep follow basket")
for bucket, count in hash_table.items():
    if(count >= 4):
        print(f"Bucket {bucket}: {count}")
    
```
